```python
# CLEANING DATA
# it is possible that some items have NaN values for some columns but have legit values for other columns so deleting any rows with NaN values could lead to loss of significant data
# therefore, depending on the attribute, we can either delete the rows entirely or keep them and manipulate the sparse values

# STEP 1: converting all columns in same format of lower case joined by _

# Creating a new DataFrame df 
df = pd.DataFrame(data)
df.columns=[col.lower().replace(" ","_") for col in df.columns]
df.columns

```

```python
# STEP 2: cleaning latitude
df["lat"].value_counts()
# dropping NaN values
df = df.dropna(subset=['lat'])
print(df.iloc[50:100,7])
check_lat = df['lat'].isnull().values.any()
# printing the result
print(check_lat)
```

```python

# STEP 3: cleaning country_code 
# replacing NaN values in country_code column to US
df['country_code'] = df['country_code'].fillna('US')
print(df.iloc[50:100,10])
check_code = df['lat'].isnull().values.any()
# printing the result
print(check_code)
```

```python

# STEP 4: cleaning room_type
df["room_type"].value_counts()
# dropping NaN values (# of missing values is neglible)
df = df.dropna(subset=['room_type'])
print(df.iloc[50:100,13])
check_room = df['room_type'].isnull().values.any()
# printing the result
print(check_room)
```

```python

# STEP 5: cleaning minimum nights
df['minimum_nights'].isna().sum()
# can either replace with 1 (because it does not make sense to have less than a night)
df['minimum_nights'] = df['minimum_nights'].fillna(1)
print(df.iloc[50:100,17])
check_nights = df['minimum_nights'].isnull().values.any()
# printing the result
print(check_nights)
```

```python
# STEP 6: cleaning reviews_per_month
print(df['reviews_per_month'].isna().sum())
# replace NaN with 0
df['reviews_per_month'] = df['reviews_per_month'].fillna(1)
print(df.iloc[50:100,20])
check_reviews = df['reviews_per_month'].isnull().values.any()
# printing the result
print(check_reviews)
```

```python
# STEP 7: cleaning name
print(df['name'].isna().sum())
# replace NaN with id's
df['name'] = df['name'].fillna(df['id'])
print(df.iloc[0:50,1])
check_name = df['name'].isnull().values.any()
# printing the result
print(check_name)
```

```python
# STEP 8: cleaning host_name
print(df['host_name'].isna().sum())
# replacing NaN values with'host'
df['host_name'] = df['host_name'].fillna('Host')
print(df.iloc[0:20,4])
check_host_name = df['host_name'].isnull().values.any()
# printing the result
print(check_host_name)
```

```python
# STEP 9: cleaning availability_365 

mean_availability_365 = 141 # as found earlier

# check for NaN, >365 and maybe <365
(df["availability_365"]>365).value_counts() # gives 2782 greater than 365
(df["availability_365"]<0).value_counts() # gives 432 less than 0
df['availability_365'].isna().sum()

print(df.iloc[0:50,23])

for row in df.index:
    # check if the row has a missing value
    if pd.isnull(df.loc[row, 'availability_365']):
        # if it does, replace it with mean value
        df.loc[row, 'availability_365'] = mean_availability_365

for row in df.index:
#    check if row has value greater than 365
    if (df.loc[row, 'availability_365']>365):
        df.loc[row, 'availability_365'] = (df.loc[row, 'availability_365'])-365
    elif (df.loc[row, 'availability_365']<365):
        df.loc[row, 'availability_365'] = abs(df.loc[row, 'availability_365'])

print(df.iloc[0:50,23])
```
